Summary

This dataset provides information of 277 retailers which services require carrier plastic bags in England for the year 2016/2017. Information in this dataset was voluntarily submitted by the retailers. Tesco is the company that issued the largest number of single use plastic bags, according to Fig. 1, among the retailes that submitted their information. Only 56% of the companies donated money from the cost obtained from the plastic bags to good causes. Out of 157 retailers that donated their proceeds 77% donated 100% or more of the money obtained. Figure 2 shows that one company donated 500% of its proceeds. This company is 'Eyre & Elliston Ltd' which had a Net proceeds of ~£10 and donated £50.
